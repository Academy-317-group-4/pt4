### Задание 1

```python
# Определяем функцию для расчета количества способов, которыми кузнечик может добраться до заданной координаты
def grasshopper_jumps(n, jumps):
    # Создаем массив для хранения количества способов добраться до каждой позиции
    F = [0] * (n + 1)
    F[0] = 1  # Начальная позиция (0) — это единственный способ остаться на месте

    # Проходим по всем позициям от 1 до n
    for i in range(1, n + 1):
        # Для каждой позиции проверяем, с каких предыдущих позиций можно до нее прыгнуть
        for jump in jumps:
            if i - jump >= 0:  # Проверяем, что прыжок не выходит за границы
                F[i] += F[i - jump]  # Увеличиваем количество способов для текущей позиции

    # Возвращаем количество способов добраться до n-й позиции
    return F

# Получаем ввод от пользователя
n = int(input("Введите n-ую координату прыжка кузнечика: "))  # Определение n-ой координаты

# Запрашиваем длины прыжков у пользователя
jumps_input = input("Введите возможные длины прыжков, разделенные пробелами (например, 1 2 3 4): ")
jumps = list(map(int, jumps_input.split()))  # Преобразуем ввод в список целых чисел

# Вызываем функцию и получаем количество способов добраться до n
F = grasshopper_jumps(n, jumps)

# Выводим результаты
print(f"Количество путей, которые позволяют дойти до конечной точки пути кузнечика: {F[n]}")
print(f"Последовательность количества путей для каждой точки: {F}")
```

### Задание 2

```python
# Определяем функцию для расчета минимальной стоимости прыжков кузнечика
def grasshopper_jumps_cost(n, jumps, costs):
    # Создаем массив для хранения минимальной стоимости до каждой позиции
    F = [float('inf')] * (n + 1)  # Инициализируем с бесконечностью
    F[0] = 0  # Стоимость на начальной позиции 0 (всё начинается с этой точки)

    # Проходим по всем позициям от 1 до n
    for i in range(1, n + 1):
        # Для каждой позиции проверяем, с каких предыдущих позиций можно до нее прыгнуть
        for j, jump in enumerate(jumps):
            if i - jump >= 0:  # Проверяем, что прыжок не выходит за границы
                F[i] = min(F[i], F[i - jump] + costs[j])  # Выбираем минимальную стоимость

    # Возвращаем минимальную стоимость, чтобы добраться до n-й позиции
    return F[n]

# Получаем ввод от пользователя
n = int(input("Введите n-ую координату прыжка кузнечика: "))  # Определение n-ой координаты

# Запрашиваем длины прыжков у пользователя
jumps_input = input("Введите возможные длины прыжков, разделенные пробелами (например, 1 2 3 4): ")
jumps = list(map(int, jumps_input.split()))  # Преобразуем ввод в список целых чисел

# Запрашиваем стоимости прыжков у пользователя
costs_input = input("Введите стоимости прыжков, разделенные пробелами (например, 1 2 3 4): ")
costs = list(map(int, costs_input.split()))  # Преобразуем ввод в список целых чисел

# Проверяем, что количество прыжков соответствует количеству стоимостей
if len(jumps) != len(costs):
    print("Ошибка: количество прыжков должно соответствовать количеству стоимостей!")
else:
    # Вызываем функцию и получаем минимальную стоимость
    min_cost = grasshopper_jumps_cost(n, jumps, costs)

    # Выводим результаты
    print(f"Минимальная стоимость, чтобы добраться до конечной точки пути кузнечика: {min_cost}")
```
